```python
import torch
import torchani
from torchani.datasets import ANIDataset
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from torchani.units import hartree2kcalmol
```

```python
df = pd.read_csv('/home/nick/CSVs/df_2x.csv')
```

```python
df_by_species = df.set_index('Species')
h_df = df_by_species.loc[1]
c_df = df_by_species.loc[6]
n_df = df_by_species.loc[7]
o_df = df_by_species.loc[8]
```

```python
h_energies = h_df['Avg_AE'].to_numpy()
h_stdev = h_df['Stdev'].to_numpy()

c_energies = c_df['Avg_AE'].to_numpy()
c_stdev = c_df['Stdev'].to_numpy()

n_energies = n_df['Avg_AE'].to_numpy()
n_stdev = n_df['Stdev'].to_numpy()

o_energies = o_df['Avg_AE'].to_numpy()
o_stdev = o_df['Stdev'].to_numpy()
```

```python
x = hartree2kcalmol(c_energies)
y = hartree2kcalmol(c_stdev)

plt.hist2d(x,y,bins=(250,250),range=[[-35,85],[-0.5,15]],cmin=1,cmap=plt.cm.gnuplot)
plt.xlabel('Predicted carbon energy contribution (kcal/mol)')
plt.ylabel('Stdev in predicted energy across ensemble (kcal/mol)')
plt.show()
```

```python
from matplotlib.ticker import PercentFormatter

%matplotlib inline
plt.rcParams['figure.dpi'] = 200
# Carbon histograms
print(len(c_stdev))



plt.hist(hartree2kcalmol(c_stdev),bins=50,weights=(np.ones(len(c_stdev)) / len(c_stdev)),color='darkviolet', range=[0,10])

plt.gca().yaxis.set_major_formatter(PercentFormatter(1,decimals=0))

#plt.axvline(x = 7.436886, color = 'r', label = 'axvline - full height')


plt.xlabel('Standard deviation in predicted carbon energy contribution (kcal/mol)')
plt.ylabel('Frequency')
#plt.title('ANI-1x first conformer subset')
plt.show()

print('Average standard deviation in carbon energy contribution:',round(hartree2kcalmol(np.mean(c_stdev)),3),'kcal/mol')
```

```python
plt.rcParams['figure.dpi'] = 200
# Nitrogen histograms
print(len(n_stdev))



plt.hist(hartree2kcalmol(n_stdev),bins=100,weights=(np.ones(len(n_stdev)) / len(n_stdev)),color='darkviolet', range=[0,15])

plt.gca().yaxis.set_major_formatter(PercentFormatter(1,decimals=0))

#plt.axvline(x = round(hartree2kcalmol(np.mean(n_stdev)),3), color = 'r', label = 'axvline - full height')


plt.xlabel('Standard deviation in predicted nitrogen energy contribution (kcal/mol)')
plt.ylabel('Frequency')
#plt.title('ANI-1x first conformer subset')
plt.show()

print('Average standard deviation in carbon nitrogen contribution:',round(hartree2kcalmol(np.mean(n_stdev)),3),'kcal/mol')
```
